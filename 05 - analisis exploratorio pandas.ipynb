https://pandas.pydata.org/pandas-docs/version/0.23/generated/pandas.core.groupby.DataFrameGroupBy.agg.html

```python
df_olympics_filtered.groupby('Name').agg({'Medal': 'count'}).sort_values(by = 'Medal', ascending = False)
```

```python
medallero = df_olympics_filtered.pivot_table(index = 'NOC', columns = 'Medal', values = 'ID', aggfunc='count', fill_value = 0)
medallero.head()
```

```python
medallero.columns
```

```python
medallero['Total'] = medallero.sum(axis = 1)
```

```python
medallero.head()
```

```python
medallero.sort_values(by='Total', ascending=False)
```

# Excercises

### What is the most successful country in the Summer Olympics?

¿How can we evaluate performance in the olympics?
* number of events?
* number of medals?

¿Is it the same to get a gold medal than a bronze medal?

```python
# correcting for duplicates

#filter only medals
df_olympics_events = df_olympics[~df_olympics.Medal.isna()]

#drop duplicates
df_olympics_events = df_olympics_events.drop_duplicates(subset = ['Games', 'Event', 'Medal'])

#pivot
medallero = df_olympics_events.pivot_table(index = 'NOC', 
                                             columns = 'Medal', 
                                             values = 'ID', 
                                             aggfunc='count', 
                                             fill_value = 0)

#add total count
medallero['Total'] = medallero.sum(axis = 1)

medallero = medallero.sort_values(by = 'Total', ascending=False)

medallero
```
